```python
import molsysmt as msm
from simtk import unit
```

# Protonation

```python
pdbfile = msm.demo_systems.files['1tcd.pdb']
```

```python
molecular_system = msm.convert(pdbfile)
```

```python
n_Hs = msm.get(molecular_system, selection='atom_type=="H"', n_atoms=True)
print('{} hydrogens in the system'.format(n_Hs))
```

```python
molecular_system = msm.add_missing_hydrogens(molecular_system, pH=7.4, engine='OpenMM')
```

```python
n_Hs = msm.get(molecular_system, selection='atom_type=="H"', n_atoms=True)
print('{} hydrogens in the system'.format(n_Hs))
```
